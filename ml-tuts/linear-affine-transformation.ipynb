```python
import numpy as np
import sys
sys.path.append('/Users/xinzheli/git_repo/SIT720-ML-Deakin/')
from my_ml_package.visualize import plot_vectors

# A geometric vector (or a line segment). Note that it's not representing a single vector in the machine learning sense.
v = np.array([[0, 0], 
              [1, 2]])

# Linear Transformation: Scaling, Rotation, Shearing (Shifts one part of an object more than another.)
scaling_matrix = np.array([[2, 0], 
                           [0, 2]])
scaled_v = v @ scaling_matrix
plot_vectors([v, scaled_v], ['blue', 'green'], 'Scaling')

theta = np.radians(45)  # 45 degree rotation
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], 
                            [np.sin(theta), np.cos(theta)]])
rotated_v = v @ rotation_matrix
plot_vectors([v, rotated_v], ['blue', 'red'], 'Rotation')

shearing_matrix = np.array([[1, 1], [0, 1]]) #  the shift occurs more in the x-direction for points along the y-axis
sheared_v = v @ shearing_matrix
plot_vectors([v, sheared_v], ['blue', 'orange'], 'Shearing')

# In machine learning, vectors often represent data points and are typically stored in row-major format where each row is a vector. 
# In contrast, in computer graphics and geometric transformations, a column-vector notation is frequently used, where vectors are columns, 
# and transformation matrices are applied to these column vectors to perform operations like translation, rotation, scaling, and shearing.
```

```python
# Affine Transformation: Combination of Rotation, Scaling and Translation
# Scaling + Rotation
theta = np.radians(45)  # 45 degree rotation
scaling_factor = 2  # Scaling by a factor of 2
rotation_scaling_matrix = np.array([[scaling_factor*np.cos(theta), -np.sin(theta)], 
                                    [np.sin(theta), scaling_factor*np.cos(theta)]])

# Apply Rotation and Scaling
transformed_v = v @ rotation_scaling_matrix

# Translation
translation_vector = np.array([1, 1])

# Apply Translation
affine_transformed_v = transformed_v + translation_vector

# Visualize
plot_vectors([v, affine_transformed_v], ['blue', 'magenta'], 'Affine Transformation')
```
