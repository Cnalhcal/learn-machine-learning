```python

# Define the algorithms
algorithms = {
    'RF': RandomForestRegressor(n_estimators=100, max_depth=None, random_state=42),
    'DT': DecisionTreeRegressor(random_state=42),
    'SVR': SVR(),
    'FFNN': MLPRegressor(random_state=42),
    'LR': LinearRegression()
}

# Define the distributions
distributions = ['Zone 1 Power Consumption', 'Zone 2  Power Consumption', 'Zone 3  Power Consumption']

# Define a DataFrame to store the results
results = pd.DataFrame(columns=['Algorithm', 'Distribution', 'RSME', 'MAE'])

# Initialize an empty list to store results
results_list = []

# Loop over each algorithm and distribution
for alg_name, model in algorithms.items():
    for dist in distributions:
        # Split features and target
        features = data.drop(columns=['DateTime', 'Temperature', 'Humidity', 'Wind Speed', 'general diffuse flows', 'diffuse flows', dist])

        target = data[dist]
        
        # Split the dataset into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
        
        # Train the model
        model.fit(X_train, y_train)
        
        # Make predictions on the test set
        predictions = model.predict(X_test)
        
        # Calculate performance metrics: RMSE and MAE
        rmse = np.sqrt(mean_squared_error(y_test, predictions))
        mae = mean_absolute_error(y_test, predictions)
        
        # Append the results to the list
        results_list.append({'Algorithm': alg_name, 'Distribution': dist, 'RSME': rmse, 'MAE': mae})

# Convert the list of dictionaries to a DataFrame
results = pd.DataFrame(results_list)

# Print the results
print(results)


```
